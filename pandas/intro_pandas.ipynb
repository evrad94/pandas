```python
# Series e DataFrame lists agguinta  con pandas
import pandas as pd # per aggiungere le funtionalità a python per manipolare data base

#Series
# Series -> prendo in ingresso mla lista di valori
# Series -> pd.series(data,index)
pds = pd.Series([1,2,3,4,5])
#print(pds)
#print(pds[2]) # per cercare un valore precisio con l'indice per esempio: [2]
print(pds[1:3]) # non è amesso l'ultimo indice

#Nelle serie è possibile esplicitare gli indici
#pdsi = pd.Series([1,2,3,4,5], index=list('abcde'))
pdsi = pd.Series([1,2,3,4,5], ['a','b','c','d','e']) #devo sempre equilibrare etichetta e indici
#print(pdsi)
#print(pdsi['c'])
print(pdsi['b':'d']) #con l'etichetta è amesso pure l'ultimo
```

```python
# Esistono due motodi diversi nelle serie che permettono 
# l'accesso a un sottoinsieme di dati : 
import pandas as pd
# loc[] -> usa le etichette è amesso l'ultimo valore
# iloc[] -> usa gli indici (slicing) non è amesso l'ultimo valore

pdsi = pd.Series([1,2,3,4,5], index=['AB', 'EF','CD',  'GH', 'IL'])

print(pdsi.iloc[3])
print(pdsi.loc["CD"])

print(pdsi.iloc[1:3])
print(pdsi.loc["AB": "CD"])
```

```python
# DataFrame
# il DataFrame è una struttura di dati Bidimensionali quindi crea una tabella ordinata 

import pandas as pd #sempre importare pandas
import numpy as np

matrix = np.random.randint(0,100, [4,5])
#print(matrix)

# posso esplicitare intestazioni di riga o di colonna
# rappresentazione del dato di un DataFrame

pd.DataFrame(matrix,
                index=['Milano', 'Torino','Roma',  'Napoli'],# index sempre righe nella tabella
                columns=['2020', '2021', '2022', '2023', '2024'])
```

```python
import pandas as pd
iris = pd.read_csv('iris.csv') # importo un CSV in un DataFrame
stats = iris.describe() # restituisce un sommario di dati statistici del dataframe
stats.to_csv('iris_stats.csv') # Salvare un dataframe in un CSV che prende un DataFrame e lo salva creando un file csv
stats.to_excel('iris_stats.xlsx',sheet_name='Foglio Stats') #Salvare un dataframe in un foglio excel
stats.to_json('iris_stats.json') # Salvare un dataframe in un formato Json
stats.to_markdown('iris_stats.md') # salva un dataframe in un formato markdown
```
